#Предобработка

```python
import pandas as pd
import numpy as np
import sqlite3
```

```python
df = pd.read_csv('/content/November_Zillow_Dallas_11-16-22.csv')
```

```python
df.head()
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
df.columns = df.columns.str.lower().str.replace(' ','_')
```

```python
df.isna().mean()
```

```python
con = sqlite3.connect('db')
```

```python
df.to_sql('house',con,index=False,if_exists='replace')
```

```python
def select(sql):
  return pd.read_sql(sql,con)
```

Узнали адреса всех домов в Dallas, их статусы, цены, спальные и вынные комнаты.  
Так же выявили самый дорогой дом в городе.

```python
sql = '''select t.addresscity, t.address, t.statustext, t.price, t.beds, t.baths from house t
where t.addresscity = 'Dallas'
order by t.beds asc
'''
```

```python
select(sql)
```

В этой таблице, по мимо предыдущих данных, мы выявили спальные комнаты по возростающей.

```python
sql = '''select t.addresscity, t.address, t.statustext, t.price, t.beds, t.baths from house t
where t.addresscity = 'Dallas'
order by t.baths asc
'''
```

```python
select(sql)
```

В этой таблице, так же как и в предыдущей, но тут вместо спальных комнат, мы выявили ванные комнаты по возростающей.

```python
sql = '''select t.addresscity, t.address, t.statustext, t.price, t.beds, t.baths, t.area from house t
where t.addresscity = 'Dallas'
order by t.area asc
'''
```

```python
select(sql)
```

В этой таблице добавили колонку - area(площадь недвижимости в квадратных футах), так же как и в предыдущих данных, посмотрим площадь по возрастающей.

```python
sql = '''select t.addresscity, t.address, t.statustext, t.price, t.beds, t.baths, t.area, t.latitude, t.longitude from house t
where t.addresscity = 'Dallas'
order by t.price asc
'''
```

```python
select(sql)
```

К этой таблице добавили - latitude и longitude(широта и долгота), то есть, узнали точные координаты и место нахождение каждого дома.

```python
sql = '''select t.addresscity, t.address, t.statustext, t.price, t.beds, t.baths, t.area, t.latitude, t.longitude, t.zestimate from house t
where t.addresscity = 'Dallas'
order by t.zestimate asc
'''
```

```python
select(sql)
```

В данной таблице добавили колонку zestimate - это компания Зиллоу, которая предоставила оценку стоимости недвижимости, так же, смотрим на цены по возрастанию.
