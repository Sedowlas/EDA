```python
df = pd.read_csv('/content/All Year Olympic Dataset (with 2020 Tokyo Olympics).csv')
```

```python
df.head()
```

```python
len(df)
```

```python
df.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
df.isna().mean()
```

```python
df.to_sql('olympic',con,index=False,if_exists='replace')
```

```python
def select(sql):
  return pd.read_sql(sql,con)
```
