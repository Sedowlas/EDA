```python
import pandas as pd
import numpy as np
```

```python
import sqlite3
```

```python
con = sqlite3.connect('db')
```

```python
df = pd.read_csv('/content/newCarsIndia.csv')
```

```python
df.head()
```

```python
df.columns = df.columns.str.lower()
```

```python
df.columns = df.columns = ['car', 'style', 'range', 'transmission', 'vehicletype', 'pricerange',
       'rating']
```

```python
df.isna().mean()
```

```python
df.to_sql('cars',con,index=False,if_exists='replace')
```

```python
def select(sql):
  return pd.read_sql(sql,con)
```

```python
pd.set_option('display.max_rows', None)
```
